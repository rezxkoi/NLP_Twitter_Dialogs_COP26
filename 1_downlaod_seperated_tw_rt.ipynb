## #COP26

```python
# Setting variables to be used in format string command below
tweet_count = 2000
text_query = "#cop26"
since_date = "2021-10-25"
until_date = "2021-11-05"
```

```python
# Using OS library to call CLI commands in Python
# Using OS library to call CLI commands in Python
os.system('snscrape --jsonl --max-results {} --since {} twitter-search "{} until:{}"> cop26.json'
          .format(tweet_count, since_date, text_query, until_date))
```

```python
# Reads the json generated from the CLI command above and creates a pandas dataframe
tweets_cop26 = pd.read_json('cop26.json', lines=True)

# Displays first 5 entries from dataframe
tweets_cop26.head()
```

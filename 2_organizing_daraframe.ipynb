## Cleaning round #1

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer, ENGLISH_STOP_WORDS
from sklearn.decomposition import NMF, TruncatedSVD, LatentDirichletAllocation
```

#### Common data cleaning steps on all text:

-Make text all lower case

-Remove punctuation

-Remove numerical values

-Remove common non-sensical text (/n)

-Tokenize text

-Remove stop words


```python
'''Make text lowercase, remove text in square brackets, remove punctuation and remove words containing numbers.'''
import re
import string

def clean_text_round1(text):
    
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
```

```python
# Apply a second round of cleaning
def clean_text_round2(text):
    '''Get rid of some additional punctuation and non-sensical text that was missed the first time around.'''
    text = re.sub('[‘’“”…]', '', text)
    text = re.sub('\n', '', text)
    return text
```

```python
clean_txt_rd1 = []
for i in range(len(df_tw['content'])):
    clean_txt_rd1.append(clean_text_round1(df_tw['content'][i]))
```

```python
df_tw['clean_rd1'] = clean_txt_rd1
```

```python
clean_txt_rd2 = []
for i in range(len(df_tw['content'])):
    clean_txt_rd2.append(clean_text_round2(df_tw['clean_rd1'][i]))
```

```python
stop_words = ENGLISH_STOP_WORDS.union(['math'])
```

```python
corpus = df_tw['clean_rd2'].fillna(value='')
```

```python
tfidf = TfidfVectorizer(stop_words=stop_words)

tweet_word_matrix = tfidf.fit_transform(corpus)

vocab = tfidf.get_feature_names()
```

# NMF

```python
nmf = NMF(n_components=5)

nmf.fit(tweet_word_matrix)
```

### Tweet-Topic Matrix

```python
tweet_topic_matrix = nmf.transform(tweet_word_matrix)
```

```python
tweet_topic_matrix
```

```python
tweet_topic_matrix_df = pd.DataFrame(tweet_topic_matrix).add_prefix('topic_')
```

```python
tweet_topic_matrix_df
```

```python
tweet_topic_matrix_df[['raw_tweets', 'clean_tweets']] = df_tw[['content', 'clean_rd2']]
tweet_topic_matrix_df.head()
```

```python
word_topic_matrix_df = pd.DataFrame(nmf.components_, columns=vocab).T.add_prefix('topic_')
word_topic_matrix_df
```

## Topic Interpretation

```python
for tweet in tweet_topic_matrix_df.sort_values(by='topic_0', ascending=False).head(10)['raw_tweets'].values:
    print(tweet)
    print()
```

```python
word_topic_matrix_df.sort_values(by='topic_0', ascending=False).head(10)
```

```python
def top_tweets(tweet_topic_matrix_df, topic, n_tweets):
    return (tweet_topic_matrix_df
            .sort_values(by=topic, ascending=False)
            .head(n_tweets)['raw_tweets']
            .values)

def top_words(word_topic_matrix_df, topic, n_words):
    return (word_topic_matrix_df
            .sort_values(by=topic, ascending=False)
            .head(n_words))[topic]
```

```python
for tweet in top_tweets(tweet_topic_matrix_df, 'topic_1', 5):
    print(tweet)
    print()
```

```python
top_words(word_topic_matrix_df, 'topic_1', 10)
```

```python
for tweet in top_tweets(tweet_topic_matrix_df, 'topic_3', 10):
    print(tweet)
    print()
```
